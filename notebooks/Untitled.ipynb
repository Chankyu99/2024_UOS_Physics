```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 파일을 읽어오는 함수
def read_pdos(file_name):
    data = np.loadtxt(file_name)
    kpoints = data[:, 0]  # k-point
    bands = data[:, 1]    # 에너지 밴드
    dos = data[:, 2]      # DOS
    return kpoints, bands, dos

# pdos 데이터 읽기
kpoints, bands, dos = read_pdos('graphene.pdos_tot')

# 플롯 생성
fig, ax1 = plt.subplots(figsize=(10, 6))

# 밴드 구조 플롯 (왼쪽 축)
ax1.plot(kpoints, bands, 'b-', label='Band Structure')
ax1.set_xlabel('K-point')
ax1.set_ylabel('Energy (eV)')
ax1.set_title('Band Structure and DOS')
ax1.grid(True)

# 오른쪽 축 설정하여 DOS 플롯
ax2 = ax1.twinx()
ax2.plot(dos, bands, 'r-', label='DOS', alpha=0.7)
ax2.set_ylabel('Density of States (DOS)')

# 범례 추가
fig.legend(loc='upper right')

# 플롯 보여주기
plt.show()
```

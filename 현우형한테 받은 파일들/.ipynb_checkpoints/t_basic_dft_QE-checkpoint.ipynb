```python
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt('./bands.dat.gnu')
dos = np.loadtxt('./pdos.dat.pdos_tot')

k = np.unique(data[:, 0])
bands = np.reshape(data[:, 1], (-1, len(k)))
plt.figure(figsize=(9,4))
plt.subplot(121)
for band in range(len(bands)):
    plt.plot(k, bands[band, :], linewidth=1, alpha=0.5, color='k')
plt.xlim(min(k), max(k))

# Fermi energy
# plt.axhline(0.921, linestyle=(0, (8, 10)), linewidth=0.75, color='k', alpha=0.5)
# High symmetry k-points (check bands_pp.out)
# plt.axvline(0.6667, linewidth=0.75, color='k', alpha=0.5)
# plt.axvline(1, linewidth=0.75, color='k', alpha=0.5)
# text labels
# plt.xticks(ticks= [0, 0.6667, 1, 1.5774], labels=['$\Gamma$', 'K', 'M', '$\Gamma$'])
plt.ylim(-9,9)
plt.ylabel("Energy (eV)")
plt.subplot(122)
plt.plot(dos[:,1],dos[:,0])
plt.ylim(-9,9)
plt.xlim(-0.1,4.5)
plt.xlabel('DOS')
plt.ylabel("Energy (eV)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
a=2.46
a1=a*np.array([1,0])
a2=a*np.array([1,0])
a3=a*np.array([np.sqrt(3)/2,1/2])
a4=a*np.array([1/2,np.sqrt(3)/2])

b1=a*np.array([1/2,np.sqrt(3)/2])
b2=a*np.array([-1/2,np.sqrt(3)/2])
b3=a*np.array([np.sqrt(3)/2,-1/2])
b4=a*np.array([1/2,-np.sqrt(3)/2])

def R(theta):
    thed2r = np.deg2rad(theta)
    R = np.array([[np.cos(thed2r),-np.sin(thed2r)],[np.sin(thed2r),np.cos(thed2r)]])
    return R
plt.figure(figsize=(9,6))
plt.subplot(221)
plt.plot((0,a1[0]),(0,a1[1]),label='a1')
plt.plot((0,b1[0]),(0,b1[1]),label='b1')
plt.scatter(1/3*(a1+b1)[0],1/3*(a1+b1)[1])
plt.scatter(2/3*(a1+b1)[0],2/3*(a1+b1)[1])
print('1_A',1/3*(a1+b1)[0],1/3*(a1+b1)[1])
print('1_B',2/3*(a1+b1)[0],2/3*(a1+b1)[1])

plt.axis('equal')
plt.subplot(222)
plt.plot((0,a2[0]),(0,a2[1]),label='a2')
plt.plot((0,b2[0]),(0,b2[1]),label='b2')
plt.scatter(((1/3)*(a2)+(2/3)*b2)[0],((1/3)*(a2)+(2/3)*b2)[1])
plt.scatter(((2/3)*(a2)+(1/3)*b2)[0],((2/3)*(a2)+(1/3)*b2)[1])
print('2_A',((1/3)*(a2)+(2/3)*b2)[0],((1/3)*(a2)+(2/3)*b2)[1])
print('2_B',((2/3)*(a2)+(1/3)*b2)[0],((2/3)*(a2)+(1/3)*b2)[1])
plt.axis('equal')
plt.subplot(223)
plt.plot((0,a3[0]),(0,a3[1]),label='a3')
plt.plot((0,b3[0]),(0,b3[1]),label='b3')
plt.scatter(1/3*(a3+b3)[0],1/3*(a3+b3)[1])
plt.scatter(2/3*(a3+b3)[0],2/3*(a3+b3)[1])
print('3_A',1/3*(a3+b3)[0],1/3*(a3+b3)[1])
print('3_B',2/3*(a3+b3)[0],2/3*(a3+b3)[1])


plt.axis('equal')
plt.subplot(224)
plt.plot((0,a4[0]),(0,a4[1]),label='a4')
plt.plot((0,b4[0]),(0,b4[1]),label='b4')
plt.scatter(((1/3)*(a4)+(2/3)*b4)[0],((1/3)*(a4)+(2/3)*b4)[1])
plt.scatter(((2/3)*(a4)+(1/3)*b4)[0],((2/3)*(a4)+(1/3)*b4)[1])
print('4_A',((1/3)*(a4)+(2/3)*b4)[0],((1/3)*(a4)+(2/3)*b4)[1])
print('4_B',((2/3)*(a4)+(1/3)*b4)[0],((2/3)*(a4)+(1/3)*b4)[1])
plt.axis('equal')
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
kb=np.loadtxt('bn_bands.dat.gnu')
knum=151
nbnd=12
c1s=np.loadtxt('bs.dat')[:,2].reshape(knum,nbnd).T.flatten()
c1pz=np.loadtxt('bpz.dat')[:,2].reshape(knum,nbnd).T.flatten()
c1px=np.loadtxt('bpx.dat')[:,2].reshape(knum,nbnd).T.flatten()
c1py=np.loadtxt('bpy.dat')[:,2].reshape(knum,nbnd).T.flatten()
c2s=np.loadtxt('ns.dat')[:,2].reshape(knum,nbnd).T.flatten()
c2pz=np.loadtxt('npz.dat')[:,2].reshape(knum,nbnd).T.flatten()
c2px=np.loadtxt('npx.dat')[:,2].reshape(knum,nbnd).T.flatten()
c2py=np.loadtxt('npy.dat')[:,2].reshape(knum,nbnd).T.flatten()
fermi=-2.3
# fermi=-0.3126
sss=15
plt.figure(figsize=(18,5))
plt.subplot(241)
plt.scatter(kb[:,0],kb[:,1]-fermi,c=c1s,s=1,cmap='Greens')
plt.title('s, Boron')
plt.xlim(0,np.max(kb[:,0]))
plt.ylim(-15,10)
plt.hlines(0,0,np.max(kb[:,0]),colors='grey',linestyles='dotted')
kpath=np.array([0,np.max(kb[:,0])/3,2*np.max(kb[:,0])/3,np.max(kb[:,0])])
plt.xticks(kpath,('$\mathrm{\Gamma}$','$\mathrm{M}$','$\mathrm{K}$','$\mathrm{\Gamma}$'),fontsize=sss)
plt.vlines(np.max(kb[:,0])/3,-15,10,colors='grey',lw=0.4)
plt.vlines(2*np.max(kb[:,0])/3,-15,10,colors='grey',lw=0.4)
plt.yticks(fontsize=sss)
plt.ylabel('Energy (eV)',fontsize=sss)
plt.colorbar()
plt.tight_layout()
plt.subplot(242)
plt.scatter(kb[:,0],kb[:,1]-fermi,c=c1pz,s=1,cmap='Greens')
plt.title('Pz, Boron')
plt.xlim(0,np.max(kb[:,0]))
plt.ylim(-15,10)
plt.hlines(0,0,np.max(kb[:,0]),colors='grey',linestyles='dotted')
kpath=np.array([0,np.max(kb[:,0])/3,2*np.max(kb[:,0])/3,np.max(kb[:,0])])
plt.xticks(kpath,('$\mathrm{\Gamma}$','$\mathrm{M}$','$\mathrm{K}$','$\mathrm{\Gamma}$'),fontsize=sss)
plt.vlines(np.max(kb[:,0])/3,-15,10,colors='grey',lw=0.4)
plt.vlines(2*np.max(kb[:,0])/3,-15,10,colors='grey',lw=0.4)
plt.yticks(fontsize=sss)
plt.ylabel('Energy (eV)',fontsize=sss)
plt.colorbar()
plt.tight_layout()
plt.subplot(243)
plt.scatter(kb[:,0],kb[:,1]-fermi,c=c1px,s=1,cmap='Greens')
plt.title('Px, Boron')
plt.xlim(0,np.max(kb[:,0]))
plt.ylim(-15,10)
plt.tight_layout()
plt.hlines(0,0,np.max(kb[:,0]),colors='grey',linestyles='dotted')
kpath=np.array([0,np.max(kb[:,0])/3,2*np.max(kb[:,0])/3,np.max(kb[:,0])])
plt.xticks(kpath,('$\mathrm{\Gamma}$','$\mathrm{M}$','$\mathrm{K}$','$\mathrm{\Gamma}$'),fontsize=sss)
plt.vlines(np.max(kb[:,0])/3,-15,10,colors='grey',lw=0.4)
plt.vlines(2*np.max(kb[:,0])/3,-15,10,colors='grey',lw=0.4)
plt.yticks(fontsize=sss)
plt.ylabel('Energy (eV)',fontsize=sss)
plt.colorbar()
plt.subplot(244)
plt.scatter(kb[:,0],kb[:,1]-fermi,c=c1py,s=1,cmap='Greens')
plt.title('Py, Boron')
plt.xlim(0,np.max(kb[:,0]))
plt.ylim(-15,10)
plt.tight_layout()
plt.hlines(0,0,np.max(kb[:,0]),colors='grey',linestyles='dotted')
kpath=np.array([0,np.max(kb[:,0])/3,2*np.max(kb[:,0])/3,np.max(kb[:,0])])
plt.xticks(kpath,('$\mathrm{\Gamma}$','$\mathrm{M}$','$\mathrm{K}$','$\mathrm{\Gamma}$'),fontsize=sss)
plt.vlines(np.max(kb[:,0])/3,-15,10,colors='grey',lw=0.4)
plt.vlines(2*np.max(kb[:,0])/3,-15,10,colors='grey',lw=0.4)
plt.yticks(fontsize=sss)
plt.ylabel('Energy (eV)',fontsize=sss)
plt.colorbar()

plt.figure(figsize=(18,5))
plt.subplot(245)
plt.scatter(kb[:,0],kb[:,1]-fermi,c=c2s,s=1,cmap='Blues')
plt.title('s, Nitrogen')
plt.xlim(0,np.max(kb[:,0]))
plt.ylim(-15,10)
plt.hlines(0,0,np.max(kb[:,0]),colors='grey',linestyles='dotted')
kpath=np.array([0,np.max(kb[:,0])/3,2*np.max(kb[:,0])/3,np.max(kb[:,0])])
plt.xticks(kpath,('$\mathrm{\Gamma}$','$\mathrm{M}$','$\mathrm{K}$','$\mathrm{\Gamma}$'),fontsize=sss)
plt.vlines(np.max(kb[:,0])/3,-15,10,colors='grey',lw=0.4)
plt.vlines(2*np.max(kb[:,0])/3,-15,10,colors='grey',lw=0.4)
plt.yticks(fontsize=sss)
plt.ylabel('Energy (eV)',fontsize=sss)
plt.colorbar()
plt.tight_layout()
plt.subplot(246)
plt.scatter(kb[:,0],kb[:,1]-fermi,c=c2pz,s=1,cmap='Blues')
plt.title('Pz, Nitrogen')
plt.xlim(0,np.max(kb[:,0]))
plt.ylim(-15,10)
plt.hlines(0,0,np.max(kb[:,0]),colors='grey',linestyles='dotted')
kpath=np.array([0,np.max(kb[:,0])/3,2*np.max(kb[:,0])/3,np.max(kb[:,0])])
plt.xticks(kpath,('$\mathrm{\Gamma}$','$\mathrm{M}$','$\mathrm{K}$','$\mathrm{\Gamma}$'),fontsize=sss)
plt.vlines(np.max(kb[:,0])/3,-15,10,colors='grey',lw=0.4)
plt.vlines(2*np.max(kb[:,0])/3,-15,10,colors='grey',lw=0.4)
plt.yticks(fontsize=sss)
plt.ylabel('Energy (eV)',fontsize=sss)
plt.colorbar()
plt.tight_layout()
plt.subplot(247)
plt.scatter(kb[:,0],kb[:,1]-fermi,c=c2px,s=1,cmap='Blues')
plt.title('Px, Nitrogen')
plt.xlim(0,np.max(kb[:,0]))
plt.ylim(-15,10)
plt.tight_layout()
plt.hlines(0,0,np.max(kb[:,0]),colors='grey',linestyles='dotted')
kpath=np.array([0,np.max(kb[:,0])/3,2*np.max(kb[:,0])/3,np.max(kb[:,0])])
plt.xticks(kpath,('$\mathrm{\Gamma}$','$\mathrm{M}$','$\mathrm{K}$','$\mathrm{\Gamma}$'),fontsize=sss)
plt.vlines(np.max(kb[:,0])/3,-15,10,colors='grey',lw=0.4)
plt.vlines(2*np.max(kb[:,0])/3,-15,10,colors='grey',lw=0.4)
plt.yticks(fontsize=sss)
plt.ylabel('Energy (eV)',fontsize=sss)
plt.colorbar()
plt.subplot(248)
plt.scatter(kb[:,0],kb[:,1]-fermi,c=c2py,s=1,cmap='Blues')
plt.title('Py, Nitrogen')
plt.xlim(0,np.max(kb[:,0]))
plt.ylim(-15,10)
plt.tight_layout()
plt.hlines(0,0,np.max(kb[:,0]),colors='grey',linestyles='dotted')
kpath=np.array([0,np.max(kb[:,0])/3,2*np.max(kb[:,0])/3,np.max(kb[:,0])])
plt.xticks(kpath,('$\mathrm{\Gamma}$','$\mathrm{M}$','$\mathrm{K}$','$\mathrm{\Gamma}$'),fontsize=sss)
plt.vlines(np.max(kb[:,0])/3,-15,10,colors='grey',lw=0.4)
plt.vlines(2*np.max(kb[:,0])/3,-15,10,colors='grey',lw=0.4)
plt.yticks(fontsize=sss)
plt.ylabel('Energy (eV)',fontsize=sss)
plt.colorbar()

plt.savefig("bn_sp.png")
plt.show()
```
